```python
# Importing Important Libraries
import numpy as np
import pandas as pd
import sklearn.datasets
```

```python
#Importing Logistic(2 Outputs) Regression Model
'''
  Two types of breast cancer

1) Benign ---> Treatable

2) Malignant ---> non Treatable
'''
from sklearn.linear_model import LogisticRegression
```

```python
# Importing accuracy score function
from sklearn.metrics import accuracy_score
```

```python
# importing dataset
breast_cancer = sklearn.datasets.load_breast_cancer()
breast_cancer
```

```python
#load the dataset into dataframe
df = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
df
```

```python
# print first 10 rows
df.head(10)
```

```python
#Adding the target (label) column to df
df['label'] = breast_cancer.target
df
```

```python
#checking the distribution of target(label) column
# 1--> benign
# 0--> malignant
df['label'].value_counts().reset_index()
```

```python
# creating a model
'''
Input Feature --> X ( All the columns except the label column)  label --> outputs are consider as labels in ML
Target Column ---> Y(Label Column)
'''
```

```python
# Seperating input feature and target column
X = df.drop(columns = 'label',axis = 1)
Y = df['label']
```

```python
# print input feature
print(X)
```

```python
# print target column or label
print(Y)
```

```python
from sklearn.model_selection import train_test_split
# splitting the data into training and testing data
X_train , X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2 ,random_state = 2)
print(X.shape, X_train.shape,X_test.shape)
print(Y.shape, Y_train.shape,Y_test.shape)
```

```python
# Training model
#  Logistic Regression is most useful for binary operation (2 outputs)
model = LogisticRegression()
# fitting the  train data into the model
model.fit(X_train,Y_train)
```

```python
# Model Evaluation
# checking the accuracy score of the training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
print(training_data_accuracy)
```

```python
# Model Evaluation
# checking the accuracy score of the testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(Y_test , X_test_prediction)
print(testing_data_accuracy)
```

```python
# Prediction
input_data = (20.57,17.77,132.9,1326,0.08474,0.07864,0.0869,0.07017,0.1812,0.05667,
              0.5435,0.7339,3.398,74.08,0.005225,0.01308,0.0186,0.0134,0.01389,0.003532,
              24.99,23.41,158.8,1956,0.1238,0.1866,0.2416,0.186,0.275,0.08902)  # 30 features only

# Convert input to numpy array
input_data_as_numpy_array = np.array(input_data)

# Reshape for prediction (1 sample, 30 features)
input_data_reshape = input_data_as_numpy_array.reshape(1, -1)

# Make prediction
prediction = model.predict(input_data_reshape)

# Interpret result
if prediction == 0:
    print("Breast cancer is Malignant")
else:
    print("Breast cancer is Benign")
```
